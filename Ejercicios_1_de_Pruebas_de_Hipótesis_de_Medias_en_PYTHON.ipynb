```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python

# Datos del problema
mu = 1  # Hipótesis nula
n = 12
x_bar = 1.4
s = 0.6
alpha = 0.05

# Cálculo del estadístico t
t_stat = (x_bar - mu) / (s / np.sqrt(n))
df = n - 1

# Valor crítico (prueba unilateral derecha)
t_critical = stats.t.ppf(1 - alpha, df)

# Valor p
p_value = 1 - stats.t.cdf(t_stat, df)

# Gráfica
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)
plt.plot(x, y, label='Distribución t-Student')

# Zona de rechazo
plt.fill_between(x, 0, y, where=(x >= t_critical), color='red', alpha=0.5, label='Zona de rechazo')

# Línea del estadístico
plt.axvline(t_stat, color='blue', linestyle='--', label=f'Estadístico t = {t_stat:.2f}')

plt.title('Prueba de hipótesis unilateral derecha (α=0.05)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de probabilidad')
plt.show()

# Resultados
print(f"Media muestral: {x_bar:.2f}")
print(f"Desviación estándar muestral: {s:.2f}")
print(f"Estadístico t: {t_stat:.4f}")
print(f"Valor p: {p_value:.4f}")

# Conclusión
if p_value < alpha:
    print("Se rechaza H0: El rendimiento promedio es superior al 1%")
else:
    print("No se rechaza H0: No hay evidencia de que el rendimiento sea superior al 1%")
```

```python


# Datos del problema
mu = 6  # Hipótesis nula
n = 30
x_bar = 5.2
s = 1.8
alpha = 0.05

# Cálculo del estadístico t
t_stat = (x_bar - mu) / (s / np.sqrt(n))
df = n - 1

# Valores críticos (prueba bilateral)
t_critical_lower = stats.t.ppf(alpha/2, df)
t_critical_upper = stats.t.ppf(1 - alpha/2, df)

# Valor p
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

# Gráfica
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)
plt.plot(x, y, label='Distribución t-Student')

# Zonas de rechazo
plt.fill_between(x, 0, y, where=(x <= t_critical_lower) | (x >= t_critical_upper),
                 color='red', alpha=0.5, label='Zona de rechazo')

# Línea del estadístico
plt.axvline(t_stat, color='blue', linestyle='--', label=f'Estadístico t = {t_stat:.2f}')

plt.title('Prueba de hipótesis bilateral (α=0.05)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de probabilidad')
plt.show()

# Resultados
print(f"Media muestral: {x_bar:.2f}")
print(f"Desviación estándar muestral: {s:.2f}")
print(f"Estadístico t: {t_stat:.4f}")
print(f"Valor p: {p_value:.4f}")

# Conclusión
if p_value < alpha:
    print("Se rechaza H0: El rendimiento es diferente al 6%")
else:
    print("No se rechaza H0: No hay evidencia de diferencia con el 6%")
```

```python


# Datos del problema
mu = 150  # Hipótesis nula
n = 36
x_bar = 147
sigma = 5  # Desviación estándar poblacional conocida
alpha = 0.01

# Cálculo del estadístico z
z_stat = (x_bar - mu) / (sigma / np.sqrt(n))

# Valores críticos (prueba bilateral)
z_critical_lower = stats.norm.ppf(alpha/2)
z_critical_upper = stats.norm.ppf(1 - alpha/2)

# Valor p
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Gráfica
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y, label='Distribución Normal')

# Zonas de rechazo
plt.fill_between(x, 0, y, where=(x <= z_critical_lower) | (x >= z_critical_upper),
                 color='red', alpha=0.5, label='Zona de rechazo')

# Línea del estadístico
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Estadístico z = {z_stat:.2f}')

plt.title('Prueba de hipótesis bilateral (α=0.01)')
plt.xlabel('Valor z')
plt.ylabel('Densidad de probabilidad')
plt.show()

# Resultados
print(f"Media muestral: {x_bar:.2f}")
print(f"Desviación estándar poblacional: {sigma:.2f}")
print(f"Estadístico z: {z_stat:.4f}")
print(f"Valor p: {p_value:.4f}")

# Conclusión
if p_value < alpha:
    print("Se rechaza H0: El precio promedio ha cambiado respecto a $150")
else:
    print("No se rechaza H0: No hay evidencia de cambio en el precio promedio")
```
