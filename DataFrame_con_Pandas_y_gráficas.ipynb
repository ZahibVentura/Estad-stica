```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### 1. Usando Pandas, hay que “leer” la base con los nombres y crear un dataframe que contenga:

a)Nombre Completo del alumno (Nombre y apellido, por lo que hay que juntar las dos columnas)

b)Edad del alumno usando la función random.randint(), con edades entre 18 y 25 años

c)Calificación del alumno usando la función random.normal() con media de 7.5 y desviación estándar de 1.0 punto

d)Porcentade Asistencia usando la función random.uniform() con mínimo de asistencia de 70% y máximo de 100%

```python
DataF = "Alumnos.csv"
#Datos  = pd.read_csv(DataF """encoding = 'latin-1'""")
Datos  = pd.read_csv(DataF)
print(Datos)
```

```python
Datos["Nombre completo"] = Datos["Nombre"] + ' ' + Datos["Apellido"]
print(Datos)
```

```python

np.random.seed(42)
Edad = [np.random.randint(18,25) for i in range(27)]
#Datos["Edad"] = np.random.randint(18,25)
Datos["Edad"] = Edad
print(Datos)
```

```python

Calificacion = [np.random.normal(7.5,1.0) for i in range(27)]
Datos["Calificación"] =Calificacion
print(Datos)
```

```python
Asistencia = [np.random.uniform(70,100) for i in range(27)]
Datos["Asistencia (%)"] = Asistencia
print(Datos)
```

###2.Imprimir el dataframe para visualizarlo, y a partir de ese DataFrame, hacer lo siguiente:

3. Usar la función de plt.hist() para hacer un histograma con la Distribución de las edades de los alumnos.

```python
plt.hist(Datos["Edad"], bins = 6, edgecolor = "k", color="red")
plt.title("Histograma de las edades")
plt.xlabel("Edades")
plt.ylabel("Frecuencia")
plt.show
```

4. Ahora hacer un gráfico de barras (plt.bar()), con dimensiones de 10 x 6 pulgadas (usar antes la función plt.figure(figsize=(10,6))) que muestre el Porcentaje de Asistencia por alumno,
en el eje de las X deben estar los nombres completos de los alumnos para cada barra y en el eje de las y el porcentaje de
asistencia iniciando desde 50% y hasta100% con plt.ylim(50).
Los nombres de los alumnos deben visualizarse verticalmente plt.xticks(rotation=90).
Usar también plt.tight_layout() para que todo quede dentro del gráfico y plt.grid(axis='y', linestyle='--', alpha=0.7)
para que se pueda ver la retícula con líneas punteadas horizontales y cierta degradación de opacidad


```python

plt.figure(figsize=(10,6))
plt.bar(Datos["Nombre completo"], Datos["Asistencia (%)"], color = "green")
plt.title("Gráfica de Asistencias")
plt.xlabel("Nombres")
plt.ylabel("% de asistencias")
plt.ylim(50)
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show
```

5. Ahora hacer un diagrama de dispersión que muestre la relación entre edad y porcentaje de asistencia usando plt.scatter(),
ponerle retícula o “grid”

```python
plt.scatter(Datos["Edad"],Datos["Asistencia (%)"], color="b", s = 10)
plt.grid(True)
plt.title("Gráfica de Asistencia - Edad")
plt.xlabel("Edad en años")
plt.ylabel("% de asistencias")
plt.show()
```

6. Hacer otro gráfico de barras con las Calificaciones de los alumnos en el mismo tamaño y sentido
del punto 4

```python
plt.figure(figsize=(10,6))
plt.bar(Datos["Nombre completo"], Datos["Calificación"], color = "orange",edgecolor = "k")
plt.title("Gráfica de Calificaciones")
plt.xlabel("Nombres")
plt.ylabel("Calificación")
plt.ylim(0)
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7, color = "k")

plt.show
```
