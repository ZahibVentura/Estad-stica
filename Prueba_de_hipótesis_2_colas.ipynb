```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Datos de la muestra
muestra = [498, 502, 495, 504, 499, 503, 497, 501, 500, 496, 505, 498]
mu_0 = 500
alpha = 0.05
```

```python


# Estadísticos muestrales
n = len(muestra)
media = np.mean(muestra)
std = np.std(muestra, ddof=1)

# Prueba t de una muestra
t_stat, p_value = stats.ttest_1samp(muestra, mu_0)

# Grados de libertad
df = n - 1

# Valor crítico para 95% (dos colas)
t_crit = stats.t.ppf(1 - alpha/2, df)
```

```python
# Mostrar resultados
print(f"Media muestral: {media:.2f}")
print(f"Desviación estándar: {std:.2f}")
print(f"t calculado: {t_stat:.4f}")
print(f"t crítico: ±{t_crit:.4f}")
print(f"p-valor: {p_value:.4f}")

if abs(t_stat) > t_crit:
    print("Se rechaza H0")
else:
    print("No se rechaza H0")
print("No hay evidencia suficiente para decir que la media de llenado es de 500ml")
```

```python
# Gráfica de la distribución t
x = np.linspace(-4, 4, 500)
y = stats.t.pdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Distribución t")
plt.fill_between(x, y, where=(x <= -t_crit), color="red", alpha=0.3, label="Zona de rechazo")
plt.fill_between(x, y, where=(x >= t_crit), color="red", alpha=0.3)
plt.axvline(t_stat, color='black', linestyle='--', label=f"t calculado = {t_stat:.4f}")
plt.title("Prueba t de dos colas (α = 0.05)")
plt.xlabel("t")
plt.ylabel("Densidad")
plt.legend()
plt.show()
```
