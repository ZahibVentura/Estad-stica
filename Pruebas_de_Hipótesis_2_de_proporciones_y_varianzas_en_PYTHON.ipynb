```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python


# Datos del problema
n = 105
x = 27
p0 = 0.35  # Hipótesis nula
alpha = 0.05

# Cálculo de la proporción muestral
p_hat = x / n

# Estadístico Z
z_stat = (p_hat - p0) / np.sqrt(p0 * (1 - p0) / n)

# Valor crítico (prueba unilateral izquierda)
z_critical = stats.norm.ppf(alpha)

# Valor p
p_value = stats.norm.cdf(z_stat)

# Gráfica
x_vals = np.linspace(-3, 3, 1000)
y_vals = stats.norm.pdf(x_vals)
plt.plot(x_vals, y_vals, label='Distribución Normal')

# Zona de rechazo
plt.fill_between(x_vals, 0, y_vals, where=(x_vals <= z_critical),
                 color='red', alpha=0.5, label='Zona de rechazo')

# Línea del estadístico
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Estadístico z = {z_stat:.2f}')

plt.title('Prueba de hipótesis unilateral izquierda (α=0.05)')
plt.xlabel('Valor z')
plt.ylabel('Densidad de probabilidad')
plt.show()

# Resultados
print(f"Proporción muestral: {p_hat:.4f}")
print(f"Estadístico z: {z_stat:.4f}")
print(f"Valor p: {p_value:.4f}")

# Conclusión
if p_value < alpha:
    print("Se rechaza H0: La proporción de comercios con pérdidas es menor al 35%")
else:
    print("No se rechaza H0: No hay evidencia de que la proporción sea menor al 35%")
```

```python

# Datos del problema
n = 200
x = 112
p0 = 0.50  # Hipótesis nula
alpha = 0.05

# Cálculo de la proporción muestral
p_hat = x / n

# Estadístico Z
z_stat = (p_hat - p0) / np.sqrt(p0 * (1 - p0) / n)

# Valores críticos (prueba bilateral)
z_critical_lower = stats.norm.ppf(alpha/2)
z_critical_upper = stats.norm.ppf(1 - alpha/2)

# Valor p
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Gráfica
x_vals = np.linspace(-3, 3, 1000)
y_vals = stats.norm.pdf(x_vals)
plt.plot(x_vals, y_vals, label='Distribución Normal')

# Zonas de rechazo
plt.fill_between(x_vals, 0, y_vals,
                 where=(x_vals <= z_critical_lower) | (x_vals >= z_critical_upper),
                 color='red', alpha=0.5, label='Zona de rechazo')

# Línea del estadístico
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Estadístico z = {z_stat:.2f}')

plt.title('Prueba de hipótesis bilateral (α=0.05)')
plt.xlabel('Valor z')
plt.ylabel('Densidad de probabilidad')
plt.show()

# Resultados
print(f"Proporción muestral: {p_hat:.4f}")
print(f"Estadístico z: {z_stat:.4f}")
print(f"Valor p: {p_value:.4f}")

# Conclusión
if p_value < alpha:
    print("Se rechaza H0: La proporción de apoyo difiere significativamente del 50%")
else:
    print("No se rechaza H0: No hay evidencia de diferencia con el 50%")
```

```python

# Datos del problema
n = 20
s2 = 0.065
sigma2_0 = 0.04  # Hipótesis nula
alpha = 0.05
df = n - 1

# Estadístico chi-cuadrado
chi2_stat = (n - 1) * s2 / sigma2_0

# Valor crítico (prueba unilateral derecha)
chi2_critical = stats.chi2.ppf(1 - alpha, df)

# Valor p
p_value = 1 - stats.chi2.cdf(chi2_stat, df)

# Gráfica
x_vals = np.linspace(0, 40, 1000)
y_vals = stats.chi2.pdf(x_vals, df)
plt.plot(x_vals, y_vals, label=f'Distribución χ² (df={df})')

# Zona de rechazo
plt.fill_between(x_vals, 0, y_vals, where=(x_vals >= chi2_critical),
                 color='red', alpha=0.5, label='Zona de rechazo')

# Línea del estadístico
plt.axvline(chi2_stat, color='blue', linestyle='--', label=f'Estadístico χ² = {chi2_stat:.2f}')

plt.title('Prueba de hipótesis unilateral derecha para varianza (α=0.05)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de probabilidad')
plt.show()

# Resultados
print(f"Varianza muestral: {s2:.4f}")
print(f"Estadístico χ²: {chi2_stat:.4f}")
print(f"Valor p: {p_value:.4f}")

# Conclusión
if p_value < alpha:
    print("Se rechaza H0: La varianza poblacional es mayor a 0.04 kg²")
else:
    print("No se rechaza H0: No hay evidencia de que la varianza sea mayor a 0.04 kg²")
```
